#Univariate Analysis

Q1 Plot Histograms using iris dataset

```

fig,ax = plt.subplots(nrows = 2, ncols = 2,figsize =(10,5) )
seplen = iris["sepal_length"]
sepwid = iris["sepal_width"]
petlen = iris["petal_length"]
petwid = iris["petal_width"]
ax[0][0].hist(seplen,bins = 20,edgecolor = 'k')
ax[0][0].set_title ("sepal_length")

ax[0][1].hist(sepwid,bins = 20,edgecolor = 'k')
ax[0][1].set_title ("sepal_width")

ax[1][0].hist(petlen,bins = 20,edgecolor = 'k')
ax[1][0].set_title ("petal_length")

ax[1][1].hist(petwid,bins = 20,edgecolor = 'k')
ax[1][1].set_title ("petal_width");

fig.tight_layout();
```

##Q13 Plot double Bar plot for the following data also add legends

```
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))
fig,ax = plt.subplots(figsize = (10,5))
ax.bar(x,men_means,width=0.4,label="Men_Means")
ax.bar(x+0.4,women_means,width = 0.4,label = "Women_Means")

ax.set_xticks(np.arange(0.2,4.4,1))
ax.set_xticklabels(labels)

for x,g,s in zip(x, men_means, women_means):
    ax.text(x-0.1, g+0.5, g)
    ax.text(x+0.3, s+0.5, s)

ax.set_title("Double bar plot")
ax.set_xlabel("LABELS")
ax.set_ylabel("Count of Means")
ax.legend(loc="best");
```

#Q20 Plot a 3D Scatterplot for iris dataset

```
fig = plt.figure(figsize = (10,7))
ax = plt.axes(projection="3d" )

x = iris['sepal_length']
y = iris['sepal_width']
z = iris['petal_length']

ax.scatter3D(x,y,z,cmap=plt.get_cmap('hsv'))

ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
ax.set_zlabel('petal_length');
```
