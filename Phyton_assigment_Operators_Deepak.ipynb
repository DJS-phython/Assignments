4. Use Boolean operators (==, !=, >, >=, <, <= ) , between a and b and Display output

```
# == , !=
a1 = int(input("Input first value to compare :"))
b1 = int(input("Input second value to compare :"))
if a1==b1:
  print ( a1, " is equal to ", b1)
elif a1!=b1:
  print (a1, " is not equal to ",b1)
```

```
# >, >=, <, <=
a1 = int(input("Input first value to compare :"))
b1 = int(input("Input second value to compare :"))
if a1>b1:
  print (a1,"is greater than" ,b1)
elif a1>=b1:
  print (a1,"is greater than or equal" ,b1)
elif a1<b1:
  print (a1,"is less than" ,b1)
elif a1<=b1:
  print (a1,"is less than or equal to " ,b1)
```

5. Create two strings x and y with any integer values stored in it
And perform bitwise operations (&,^, | ,>>,<<)

```
a=5
b=10

print (a & b) # Bitwise AND (&) operation = Both value needs to be true to get o/p as true (1)
print (a ^ b) # Bitwise XOR (^) Operation - Any one of the value needs to be True to get output as true (1)
print (a | b) # Bitwise OR (|) - Any one of the value needs to be True to get output as true (1)
print (a >> 1) # Bitwise right shift (>>) - Shift the binary value to right by one. 
print (a << 1) # Bitwise left shift (<<) - Shift the binary value to left by one.
```
